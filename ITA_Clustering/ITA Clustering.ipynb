```python
import pandas as pd
import numpy as np


from matplotlib import pyplot as plt
```

# Data Set Importing

```python
ITA_df= pd.read_csv("ITA Map Data.csv")
```

# Data Set Analysing

```python
ITA_df.describe().T
```

```python
from scipy.stats import zscore

ITA_Parmeters = ITA_df.drop('UUT NAME', axis=1)

Application_name= ITA_df.pop('UUT NAME')

ITA_Parmeters=ITA_Parmeters.apply(zscore)
ITA_df_z= ITA_Parmeters.join(Application_name)
```

```python
import seaborn as sns
sns.pairplot(ITA_Parmeters, diag_kind='kde')
```

# Building KMeans Clustering Model

```python
from sklearn.cluster import KMeans
```

```python
ITA_df_z=ITA_df_z.drop('UUT NAME', axis=1)
```

# Predicting the ITA Cluster by Application Wise 

```python
ITA_Cluster_Modle = KMeans(n_clusters=7, random_state=1)

ITA_Cluster_Modle.fit(ITA_df_z)
ITA_Cluster_labels=ITA_Cluster_Modle.labels_
Predictions= ITA_Cluster_Modle.predict(ITA_df_z)

ITA_df["ITA_Cluster"]=Predictions
ITA_df["UUT Name"]=Application_name
ITA_df[["UUT Name", "ITA_Cluster"]]
```

```python
cluster_range = range(1,10)
for i in cluster_range:
    points= np.array(ITA_df.loc[ITA_df['ITA_Cluster']==i])
    
points
```

```python
colors = ['r', 'g', 'b', 'y','r','c','m']

fig, ax = plt.subplots()

c= np.array([3,3])

for i in ITA_Cluster_labels:
    
    points = np.array(ITA_df.loc[ITA_df["ITA_Cluster"]==i])
    ax.scatter(points[:,0],points[:,1], c=colors[i])
    
```
