Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference. (5 points)
Goal_diff_count = GoalsFor - GoalsAgainst

```python
def Goal_diff_count (goal_diff, req='Goal_diff'):
    
    goal_diff['GoalsFor']=pd.to_numeric(goal_diff['GoalsFor'])
    goal_diff['GoalsAgainst']=pd.to_numeric(goal_diff['GoalsAgainst'])
    
    goal_diff['Goal_diff_count']=goal_diff['GoalsFor']-goal_diff['GoalsAgainst']
    
    if (req=='Goal_diff'):
        return goal_diff.loc[:,['Team','Goal_diff_count']]
    elif(req=="MaxMin"):
        max = goal_diff[goal_diff['Goal_diff_count']==goal_diff['Goal_diff_count'].max()]['Team']
        min = goal_diff[goal_diff['Goal_diff_count']==goal_diff['Goal_diff_count'].min()]['Team']
        return max, min
```

```python
team_df=team_df.replace("-", 0)
Goal_diff_count(team_df)
```

```python
Diff_Max, Diff_min= Goal_diff_count(team_df, "MaxMin")

print(f"Max goal Diff \n\t {Diff_Max} \n\t Min Goal Diff \n\t {Diff_min}")
```

Create a new column with name “Winning Percent” and append it to the data set (5 points)
Percentage of Winning = (GamesWon / GamesPlayed)*100

If there are any numerical error, replace it with 0%

Print the top 5 teams which has the highest Winning percentage

```python
# Converting 'GamesWon' and 'GamesPlayed' to numeric data type
# Calculating Percentage of wining,where PercantageOfWinning = (GamesWon/GamesPlayed)*100
# And restricting the decimal places to 3 
team_df["Winning Percent (in %)"]= round((pd.to_numeric(team_df['GamesWon']))/(pd.to_numeric(team_df['GamesPlayed']))*100,3)
# For 0/0 we get Nan so replacing that Nan with 0 using fillna method
team_df.fillna(0)
team_df.sort_values(by='Winning Percent (in %)', ascending=False).loc[:4,['Team','Winning Percent (in %)']]
```
